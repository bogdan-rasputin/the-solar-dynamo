```python
import numpy as np
import matplotlib.pyplot as plt

from scipy import integrate
from pathlib import Path
```

```python
def f(x, y):
    return x**2 + x*y - 2*x

def g(x, y):
    return x*y**2 + 3*y
```

```python
plt.close('all')
coords = np.linspace(-5, 5, 30)
x, y = np.meshgrid(coords, coords)
dx = x**2 + x*y - 2*x
dy = x*y**2 + 3*y
plt.figure(figsize=(12,12))
plt.gca().set_aspect('equal', adjustable='box')  # Make plot box square
plt.xlabel('$x$')
plt.ylabel('$y$')
#plt.title('2D Vector field: $f(x,y)=(-1-x^2+y, 1+x-y^2)$')
plt.quiver(x, y, dx, dy)  # plot field as quiver
plt.streamplot(x, y, dx, dy)  # plot streamlines of field.
#plt.savefig('G3-DHO-underdampened.svg')

# plotting nullclines here

# y = 2 - x

x_vals = coords
y_vals = 2 - x_vals

plt.plot(x_vals, y_vals, color='green', linestyle='--', label='x-isocline')

y1_vals = -3/x_vals

plt.plot(x_vals, y1_vals, color='red', linestyle='--', label='y-isocline')

plt.axhline(y=0, color='red', linestyle='-.', label='y-isocline')
plt.axvline(x=0, color='green', linestyle='-.', label='x-isocline')

plt.xlim(-5,5)
plt.ylim(-5,5)

fixed_points = np.array([[0, 0], [3, -1], [-1, 3]])
plt.plot(fixed_points[:,0], fixed_points[:,1], 'ko', label='Fixed Points')

plt.savefig('1a.svg')

plt.show()
```
